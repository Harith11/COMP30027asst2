```python
import pandas as pd

df = pd.read_csv('test_dataset.csv')
df.describe()
```

```python
# Apply frequency encoding to string columns with names and drop the initial columns
categorical_columns = ['director_name', 'actor_2_name', 'actor_1_name', 'actor_3_name']
for col in categorical_columns:
    frequency_encoding = df[col].value_counts(normalize=True).to_dict()
    df[col + '_freq'] = df[col].map(frequency_encoding)
    df.drop(col, axis=1, inplace=True)  # Drop the initial column

df.head()
```

```python
import pandas as pd
import numpy as np

# Define a function to convert the string of numbers into a list of floats
def convert_to_floats(embedding_str):
    numbers_str = embedding_str.strip('[]')
    numbers_str_list = list(filter(None, numbers_str.split(' ')))
    return [float(num) for num in numbers_str_list]

df['title_embedding'] = df['title_embedding'].apply(convert_to_floats)

embedding_df = pd.DataFrame(df['title_embedding'].tolist(), columns=[f'title_embedding_{i}' for i in range(len(df['title_embedding'].iloc[0]))])

df = pd.concat([df, embedding_df], axis=1)

df.drop('title_embedding', axis=1, inplace=True)
df.drop('movie_title', axis=1, inplace=True)

df.head()
```

```python
from gensim.models import Word2Vec
import pandas as pd

df['genres_list'] = df['genres'].str.split('|')
df['plot_keywords_list'] = df['plot_keywords'].str.split('|')

# Train separate Word2Vec models for 'genres' and 'plot_keywords'
model_genres = Word2Vec(sentences=df['genres_list'], vector_size=50, window=5, min_count=1, workers=4)
model_plot_keywords = Word2Vec(sentences=df['plot_keywords_list'], vector_size=50, window=5, min_count=1, workers=4)

vectors_genres = [model_genres.wv[genres].mean(axis=0) for genres in df['genres_list']]
vectors_plot_keywords = [model_plot_keywords.wv[keywords].mean(axis=0) for keywords in df['plot_keywords_list']]

df_vectors_genres = pd.DataFrame(vectors_genres, columns=[f'genres_{i}' for i in range(50)])
df_vectors_plot_keywords = pd.DataFrame(vectors_plot_keywords, columns=[f'plot_keywords_{i}' for i in range(50)])

df_vectors_genres.shape, df_vectors_genres.head(), df_vectors_plot_keywords.shape, df_vectors_plot_keywords.head()
```

```python
df = pd.concat([df, df_vectors_genres, df_vectors_plot_keywords], axis=1)

df.drop(['genres', 'plot_keywords', 'genres_list', 'plot_keywords_list'], axis=1, inplace=True)

df.shape, df.head()
```

```python
# Apply one-hot encoding to the 'content_rating' column
content_rating_dummies = pd.get_dummies(df['content_rating'], prefix='content_rating', drop_first=True)

df = pd.concat([df, content_rating_dummies], axis=1)

df.drop('content_rating', axis=1, inplace=True)

df.head()
```

```python
# Apply one-hot encoding to the 'country' and 'language' columns
country_dummies = pd.get_dummies(df['country'], prefix='country')
language_dummies = pd.get_dummies(df['language'], prefix='language')

df= pd.concat([df, country_dummies, language_dummies], axis=1)

df.drop(['country', 'language'], axis=1, inplace=True)

df.head()
```
